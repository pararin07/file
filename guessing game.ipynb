```python
import random 
name = input("what is your name:")

print("good luck", name, "!")

words = ["hello","rain","computer","guess","science","up","down","math","nerd"]
#list of words user must guess from 

word = random.choice(words)
#system will give us a random word

print("guess the characters")

guesses = '' 

turns = 20

while turns > 0:
    #counts how many turns have been used

    failed = 0 

    for character in word:
        if character in guesses:
            print(character, end='')

        else:
            print("__")
            failed += 1 
        

        
    if failed == 0:
        print('/n')
        print("winner!!!")
        print("the word is:", word)
        break

    print()
    guess = input('guess a character:')
    guesses += guess
    if guess not in word:
        turns -= 1
        print("wrong")
        print(+turns, "more guesses left")

        if turns  == 0:
            print("YOU LOSE")




```
